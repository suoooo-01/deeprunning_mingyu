## 3.2.3 계단 함수의 그래프

```python
import numpy as np 
import matplotlib.pylab as plt
def step_function(x): 
    return np.array(x > 0, dtype=int)
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # Y축의 범위 지정
plt.show()
```

## 3.2.4 시그모이드 함수 구현하기

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x= np.array([-1.0, 1.0, 2.0])
print(sigmoid(x))

t= np.array ([1.0, 2.0, 3.0])
print(1.0 + t)
print(1.0 / t)

x= np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x) # sigmoid() 함수로 변경
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y축 범위 지정
plt.show()
```
